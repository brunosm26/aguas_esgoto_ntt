```python
from pyspark.sql.functions import col, count

# Ler Silver
df = spark.read.format("delta").load(
    "/Volumes/workspace/default/silver/corte_agua/"
)
```

```python
if "dia_semana_nome" not in df.columns:
    raise Exception("A coluna 'dia_semana_nome' não existe na Silver!")

df_dia_semana = (
    df.groupBy("dia_semana_nome")
      .agg(count("*").alias("qtd_cortes"))
      .orderBy(col("qtd_cortes").desc())
)

display(df_dia_semana)
```

```python
df_dia_semana.write.format("delta") \
    .mode("overwrite") \
    .save("/Volumes/workspace/default/gold/corte_agua/cortes_por_dia_semana/")

print("Gold 03 — Cortes por dia da semana gerado com sucesso!")
```
