```python
from pyspark.sql.functions import col

# Ler GOLD por bairro existente
df_bairro = spark.read.format("delta").load(
    "/Volumes/workspace/default/gold/corte_agua/cortes_por_bairro/"
)

display(df_bairro)

# Criar TOP 10
df_top10 = (
    df_bairro
        .orderBy(col("qtd_cortes").desc())
        .limit(10)
)

display(df_top10)

# Salvar GOLD Top 10
df_top10.write.format("delta") \
    .mode("overwrite") \
    .save("/Volumes/workspace/default/gold/corte_agua/top10_bairros/")

print("GOLD TOP 10 BAIRROS gerado com sucesso!")
```
