```python
%sql
CREATE OR REPLACE TABLE classes.silver.consumo_agua_clean AS
SELECT DISTINCT
    TRIM(`Região`) AS `regiao`,
    TRIM(`Cidade`) AS `cidade`,
    TRIM(`Localidade`) AS `localidade`,
    COALESCE(`Consumo por m³`, 0) AS `consumo_por_m3`,
    REPLACE(`Valor da Conta`, 'R$', '') AS `valor_conta`,
    COALESCE(CAST(REPLACE(`Valor da Conta`, 'R$', '') AS DECIMAL(10,2)), 0) AS `valor_conta_numeric`,
    COALESCE(`Valor de Imposto`, 0) AS `valor_imposto`,
    TRIM(`Nome do Cliente`) AS `nome_cliente`,
    TRIM(`Mês de Referência`) AS `mes_referencia`,
    `Data de Vencimento` AS `data_vencimento`  
FROM classes.bronze.meu_csv_raw;
```

```python
%sql
-- Consultando as primeiras 10 linhas da tabela 'consumo_agua_clean' na camada silver
SELECT * FROM classes.silver.consumo_agua_clean LIMIT 10;
```
