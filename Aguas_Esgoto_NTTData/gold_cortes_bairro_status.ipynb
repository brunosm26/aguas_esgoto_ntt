```python
from pyspark.sql import functions as F

# Carregar Silver
df_silver = spark.read.format("delta").load("/Volumes/workspace/default/silver/corte_agua/")
```

```python
gold_07 = (
    df_silver.groupBy("bairro", "status_do_corte")
             .agg(F.count("*").alias("total_cortes"))
             .orderBy("bairro", "status_do_corte")
)
```

```python
gold_07.write.format("delta").mode("overwrite").save("/Volumes/workspace/default/gold/corte_agua/gold_07/")

display(gold_07)
```

```python
%sql
CREATE TABLE IF NOT EXISTS gold_corte_agua_gold_07 AS
SELECT *
FROM delta.`/Volumes/workspace/default/gold/corte_agua/gold_07`;

```
