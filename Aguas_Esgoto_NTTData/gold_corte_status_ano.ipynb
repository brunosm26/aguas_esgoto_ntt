```python
from pyspark.sql import functions as F

# Carrega a Silver
df_silver = spark.read.format("delta").load("/Volumes/workspace/default/silver/corte_agua/")
```

```python
gold_06 = (
    df_silver.groupBy("ano_corte", "status_do_corte")
             .agg(F.count("*").alias("total_cortes"))
             .orderBy("ano_corte", "status_do_corte")
)
```

```python
gold_06.write.format("delta").mode("overwrite").save("/Volumes/workspace/default/gold/corte_agua/gold_06/")

display(gold_06)
```

```python
%sql
CREATE TABLE IF NOT EXISTS gold_corte_agua_gold_06 AS
SELECT *
FROM delta.`/Volumes/workspace/default/gold/corte_agua/gold_06`;

```
