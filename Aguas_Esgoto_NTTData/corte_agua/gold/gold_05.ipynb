```python
from pyspark.sql import functions as F

# Carregar Silver
df_silver = spark.read.format("delta").load("/Volumes/workspace/default/silver/corte_agua/")
```

```python
# GOLD 05 â€” Ranking Top 10 bairros
gold_05 = (
    df_silver.groupBy("bairro")
             .agg(F.count("*").alias("total_cortes"))
             .orderBy(F.col("total_cortes").desc())
             .limit(10)
)

display(gold_05)
```

```python
gold_05.write.format("delta") \
    .mode("overwrite") \
    .save("/Volumes/workspace/default/gold/corte_agua/ranking_top_bairros/")
```

```python
%sql
CREATE TABLE IF NOT EXISTS gold_corte_agua_gold_05 AS
SELECT *
FROM delta.`/Volumes/workspace/default/gold/corte_agua/ranking_top_bairros`;
```
