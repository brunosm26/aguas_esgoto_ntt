```python
df_raw = (
    spark.read
        .format("csv")
        .option("header", "true")
        .option("delimiter", ",")
        .option("inferSchema", "true")
        .load("/Volumes/workspace/default/bronze/corte_agua.csv")
)

display(df_raw)
 
```

```python
df_bronze = df_raw.toDF(*[
    c.lower()
     .replace(" ", "_")
     .replace("ã", "a")
     .replace("á", "a")
     .replace("â", "a")
     .replace("ç", "c")
     .replace("é", "e")
     .replace("ê", "e")
     .replace("í", "i")
     .replace("ó", "o")
     .replace("ô", "o")
     .replace("ú", "u")
     .replace("ã", "a")
     .replace("õ", "o")
     for c in df_raw.columns
])

print("Colunas finais:")
print(df_bronze.columns)

display(df_bronze)

# ------------------------------------------
# Salvando Bronze como Delta
# ------------------------------------------

df_bronze.write.format("delta") \
    .mode("overwrite") \
    .save("/Volumes/workspace/default/bronze/corte_agua_delta/")

print("BRONZE criado com sucesso!")
```
